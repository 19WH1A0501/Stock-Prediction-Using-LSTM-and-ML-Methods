```python
barWidth = 0.25
fig = plt.subplots(figsize=(8, 5))

without_twitter = [8.33, 9.8, 9.56, 18.3]
with_twitter = [7.27, 6.43, 9.59, 18.4]

br1 = np.arange(len(without_twitter))
br2 = [x + barWidth for x in br1]

# Make the plot
plt.bar(br1, without_twitter, color=(0.2, 0.4, 0.6), width=barWidth,
        edgecolor='grey', label='without_twitter')
plt.bar(br2, with_twitter, color=(0.7, 0.4, 0.2), width=barWidth,
        edgecolor='grey', label='with_twitter')

for i, v in enumerate(without_twitter):
    plt.text(i - 0.13, v + 0.5, str(v), color='black')
for i, v in enumerate(with_twitter):
    plt.text(i + 0.17, v + 0.5, str(v), color='black')

plt.ylim(0, 20)

plt.title("Company: WIPRO (Stock price range: (162, 721))", fontweight='bold')
plt.xlabel('Model', fontweight='bold', fontsize=10)
plt.ylabel('RMSE Value', fontweight='bold', fontsize=10)
plt.xticks([r + 0.15 for r in range(len(without_twitter))],
           ['CNN-LSTM', 'Random Forest', 'SVM', 'XGBoost'])

plt.legend()
plt.show()
```

```python
tweets_analysis = []
for tweet in tweets:
    try:
        sentiment_result = analysis(tweet)[0]
        top_sentiment = max(sentiment_result, key=lambda x: x['score']) # Get the sentiment with the higher score
        tweets_analysis.append({'tweet': tweet, 'sentiment': top_sentiment['label']})
 
    except Exception as e:
        print(e)
```

```python
import pandas as pd
 
# Load the data in a dataframe
pd.set_option('max_colwidth', None)
pd.set_option('display.width', 3000)
df = pd.DataFrame(tweets_analysis)
 
# Show a tweet for each sentiment
display(df[df["sentiment"] == 'Positive'].head(1))
display(df[df["sentiment"] == 'Neutral'].head(1))
display(df[df["sentiment"] == 'Negative'].head(1))
```
